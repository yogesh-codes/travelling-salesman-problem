```python
import matplotlib.pyplot as plt


def plot_tsp_solution(cities, route):
    # Unpack the city coordinates
    x_coords, y_coords = zip(*[cities[i] for i in route])

    # Plot the cities as points
    plt.scatter(x_coords, y_coords)
    
    # Plot the path
    plt.plot(x_coords, y_coords, '-o')
    
    # Optionally, you can annotate the cities with their indices
    for i, (x, y) in enumerate(cities):
        plt.text(x, y, f'{i}')
    
    plt.title("TSP Route")
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.show()
    
```

```python
import random

def generate_random_cities(n, x_max=100, y_max=100):
    return [(random.randint(0, x_max), random.randint(0, y_max)) for _ in range(n)]

# Generate 10 random cities
cities = generate_random_cities(10)
```

```python
# Assuming 'route' is the output of your TSP solver, e.g., [0, 2, 3, 1, 0] for 4 cities
route = [0, 2, 3, 5,1,6,7,8,9, 0]  # Example route, replace with your algorithm's output

plot_tsp_solution(cities, route)
```

```python
%matplotlib inline
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Example city coordinates
cities = np.array([(np.random.uniform(0, 100), np.random.uniform(0, 100)) for _ in range(10)])

# Function to simulate algorithm's route generation for demonstration
def generate_routes(cities):
    np.random.shuffle(cities)
    return cities

# Preparing the plot
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'ro-', animated=True)

def init():
    ax.set_xlim(0, 100)
    ax.set_ylim(0, 100)
    return ln,

def update(frame):
    # Simulating route update - replace this with actual algorithm steps
    new_route = generate_routes(cities)
    xdata = new_route[:, 0]
    ydata = new_route[:, 1]
    ln.set_data(xdata, ydata)
    return ln,

# Creating the animation
ani = FuncAnimation(fig, update, frames=range(100), init_func=init, blit=True)

plt.show()
```
